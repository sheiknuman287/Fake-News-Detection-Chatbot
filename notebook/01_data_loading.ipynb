```python
import pandas as pd

fake_df = pd.read_csv("../data/Fake.csv", encoding="latin-1")
true_df = pd.read_csv("../data/True.csv", encoding="latin-1")

print(fake_df.shape)
print(true_df.shape)

fake_df.head(2)
```

```python
# Keep only the useful columns
fake_df = fake_df[['title', 'text', 'subject', 'date']]
true_df = true_df[['title', 'text', 'subject', 'date']]

print(fake_df.shape)
print(true_df.shape)
```

```python
fake_df['label'] = 0
true_df['label'] = 1
```

```python
df = pd.concat([fake_df, true_df], axis=0)
df = df.sample(frac=1).reset_index(drop=True)  # shuffle

print(df.shape)
df.head(2)
```
